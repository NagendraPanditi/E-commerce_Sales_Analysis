```python
  #Importing the Pandas Library and connect to the MySql 
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Data Loading
df = pd.read_csv("Ecommerce_Sales_Data.csv")
df.head()
```

```python
#Checking the Null_Values
df.isnull().sum() 
```

```python
#Changing the columns into Snake Case format for Code Consistency, easy to read, and easy to reference the code in sql and python without running into errors)
df = df.rename(columns = {"category" : "Product_Category"} )
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.head()
```

```python
#sorting the data in Ascending Order(Low Value to High Value)
df.sort_values(by = "order_date", ascending = True , inplace = True) 
df.head(10)
```

```python
df.info()
df.head(10)
# df.to_csv("Ecommerce_Sales_Data.csv", index = False)
```

```python
#Load the data into SQL 
#Creating Engine 
engine = create_engine("mysql+mysqlconnector://root:Naagendra%40097@localhost:3306/Sales_Database")

#Create a Table to store the data which is cleanned
table_name = "Sale_Data_Table"
df.to_sql(table_name, con=engine, if_exists='replace', index = False)

print(f" Data Succefully loaded into MySql ")

```
